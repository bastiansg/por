```python
import os
import qrcode

from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers.pil import GappedSquareModuleDrawer

from imagekitio import ImageKit
from imagekitio.models.UploadFileRequestOptions import UploadFileRequestOptions
```

```python
IMAGEKIT_PUBLIC_KEY = os.getenv("IMAGEKIT_PUBLIC_KEY")
IMAGEKIT_PRIVATE_KEY = os.getenv("IMAGEKIT_PRIVATE_KEY")
```

```python
url_endpoint = "https://ik.imagekit.io/ddmoon"
imagekit = ImageKit(
    public_key=IMAGEKIT_PUBLIC_KEY,
    private_key=IMAGEKIT_PRIVATE_KEY,
    url_endpoint=url_endpoint,
)
```

```python
options = UploadFileRequestOptions(
    folder="/istm",
    is_private_file=False,
)

image_path = "/resources/images/f69a454537a24ee1aebc05f197186258-gen.jpg"
with open(image_path, "rb") as image_file:
    upload = imagekit.upload_file(
        file=image_file,
        file_name="test.jpg",
        options=options,
    )
```

```python
image_url = f"{url_endpoint}{upload.file_path}"
qr = qrcode.QRCode(
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=34,
    border=1,
)

qr.add_data(image_url)
qr.make(fit=True)

img = qr.make_image(
    image_factory=StyledPilImage,
    module_drawer=GappedSquareModuleDrawer(size_ratio=0.9),
    fill_color="black",
    back_color="white",
)

qr_pil_image = img.get_image()
qr_pil_image
```

```python
print(qr_pil_image.size)
qr_pil_image.save("/resources/images/qr-test.jpg")
```
