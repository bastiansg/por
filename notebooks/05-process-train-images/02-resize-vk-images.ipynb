```python
import os

from tqdm import tqdm
from glob import glob
from PIL import Image
from common.utils.path import create_path
from torchvision.transforms import Resize
```

```python
resize = Resize(
    size=768,
    max_size=2000,
)
```

```python
image_paths = glob("/resources/vk-ilustrations/*.jpeg")
print(f"image_paths => {len(image_paths)}")
```

```python
output_path = "/resources/train-vk-images"
create_path(path=output_path)

for image_path in tqdm(image_paths):
    image = Image.open(image_path)
    resized_image = resize(image)
    out_image_path = f"{output_path}/{os.path.basename(image_path)}"
    resized_image.save(out_image_path)
```
