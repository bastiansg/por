```python
import os

from tqdm import tqdm
from glob import glob
from PIL import Image
from common.utils.path import create_path
from torchvision.transforms import Resize
```

```python
left_crop = 140
right_crop = 149
top_crop = 134
bottom_crop = 478

resize = Resize(
    size=768,
    max_size=2000,
)
```

```python
def process_image(
    image: Image.Image,
    resize: bool = False,
) -> Image.Image:
    width, height = image.size
    croped_image = image.crop(
        (
            left_crop,
            top_crop,
            width - right_crop,
            height - bottom_crop,
        )
    )

    if resize:
        return resize(croped_image)

    return croped_image
```

```python
image_paths = glob("/resources/train-images/*.jpg")
print(f"image_paths => {len(image_paths)}")
```

```python
output_path = "/resources/cropped-train-images"
# output_path = "/resources/resized-images"

create_path(path=output_path)
for image_path in tqdm(image_paths):
    image = Image.open(image_path)
    processed_image = process_image(image=image)
    # processed_image = process_image(image=image, resize=True)
    out_image_path = f"{output_path}/{os.path.basename(image_path)}"
    processed_image.save(out_image_path)
```
