```python
import os

from tqdm import tqdm
from glob import glob
from PIL import Image
from common.utils.path import create_path
```

```python
left_crop = 140
right_crop = 149
top_crop = 134
bottom_crop = 478
```

```python
def crop_image(image: Image.Image) -> Image.Image:
    width, height = image.size
    return image.crop(
        (
            left_crop,
            top_crop,
            width - right_crop,
            height - bottom_crop,
        )
    )
```

```python
image_paths = glob("/resources/train-images/*.jpg")
print(f"image_paths => {len(image_paths)}")
```

```python
output_path = "/resources/cropped-train-images"
create_path(path=output_path)

for image_path in tqdm(image_paths):
    image = Image.open(image_path)
    cropped_image = crop_image(image=image)
    out_image_path = f"{output_path}/{os.path.basename(image_path)}"
    cropped_image.save(out_image_path)
```
