```python
from pathlib import Path

from glob import glob
from PIL import Image
from tqdm import tqdm
```

```python
image_paths = glob("/resources/images/malika-favre/*")
print(f"image_paths: {len(image_paths)}")
```

```python
out_path = "/resources/images/malika-favre-jpg"
for image_path in tqdm(image_paths, ascii=True):
    img = Image.open(image_path)
    img = img.convert("RGB")

    file_name = Path(image_path).stem
    img.save(
        f"{out_path}/{file_name}.jpg",
        "JPEG",
        quality=100,
    )
```
