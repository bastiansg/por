```python
from glob import glob
from PIL import Image
from tqdm import tqdm
```

```python
image_paths = glob("/resources/images/malika-favre/*")
print(f"image_paths: {len(image_paths)}")
```

```python
out_path = "/resources/images/malika-favre-jpg"
for idx, image_path in tqdm(enumerate(image_paths), ascii=True):
    img = Image.open(image_path)
    img = img.convert("RGB")

    img.save(
        f"{out_path}/{idx}.jpg",
        "JPEG",
        quality=100,
    )
```
