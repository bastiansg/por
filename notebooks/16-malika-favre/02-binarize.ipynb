```python
from glob import glob
from tqdm import tqdm
from pathlib import Path

from por.utils.image import binarize
```

```python
image_paths = glob("/resources/images/malika-favre-jpg/*")
print(f"image_paths: {len(image_paths)}")
```

```python
out_path = "/resources/images/malika-favre-bw"
for image_path in tqdm(image_paths, ascii=True):
    bw_image = binarize(image_path=image_path)

    file_name = Path(image_path).stem
    bw_image.save(f"{out_path}/{file_name}.jpg")
```
