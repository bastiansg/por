```python
from PIL import Image
from escpos.printer import Usb
from torchvision import transforms
```

```python
image = Image.open("/resources/images/full-moon.jpg").convert("RGB")
resize_transform = transforms.Resize(size=400)
image = resize_transform(image)
image
```

```python
image_path = "/tmp/moon.jpg"
image.save(image_path)
```

```python
printer = Usb(
    0x04B8,
    0x0E27,
    0,
    profile="TM-T20II",
)
```
