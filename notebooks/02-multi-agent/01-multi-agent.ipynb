```python
import uuid
import logfire

from rich.pretty import pprint
from por.multi_agent import get_multi_agent, get_multi_agent_config
```

```python
logfire.configure(service_name="por")
_ = logfire.instrument_pydantic_ai()
_ = logfire.instrument_openai()
```

```python
multi_agent = get_multi_agent()
multi_agent.display_graph(
    # draw_type="ascii",
    draw_type="mermaid",
)
```

```python
multi_agent_config = get_multi_agent_config(model="grcra")
# pprint(multi_agent_config)
```

```python
image_id = uuid.uuid4().hex
state = await multi_agent.run(
    input_state={
        "image_id": image_id,
    },
    config=multi_agent_config,
    thread_id=image_id,
)

pprint(state)
print(state["image_generation_prompt"]["prompt"])
print(state["image_url"])
```
