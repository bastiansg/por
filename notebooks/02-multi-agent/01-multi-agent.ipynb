```python
import uuid

from rich.pretty import pprint
from istm.multi_agent import get_multi_agent, get_multi_agent_config
```

```python
multi_agent = get_multi_agent()
multi_agent.display_graph()
```

```python
multi_agent_config = get_multi_agent_config(model="grcra")
pprint(multi_agent_config)
```

```python
image_id = uuid.uuid4().hex
state = await multi_agent.run(
    input_state={
        "image_id": image_id,
    },
    config=multi_agent_config,
    thread_id=image_id,
)

pprint(state)
```
