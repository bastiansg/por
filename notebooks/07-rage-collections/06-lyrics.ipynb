```python
import random

from por.loaders import LyricsLoader

from rich.pretty import pprint
from rich.markdown import Markdown

from rage.retriever import Retriever
from rage.splitters import DocumentSplitter
from rage.utils.embeddings import get_openai_embeddings
```

```python
loader = LyricsLoader()
splitter = DocumentSplitter()
retriever = Retriever(dense_embeddings=get_openai_embeddings())
```

```python
documents = await loader.load(
    source_path="/resources/documents/lyrics/lyrics.json"
)

text_chunks = splitter.split_documents(documents=documents)
print(f"text_chunks => {len(text_chunks)}")
```

```python
text_chunk = random.choice(text_chunks)

pprint(text_chunk.metadata)
Markdown(text_chunk.text)
```

```python
collection_name = "lyrics"
retriever.create_collection(collection_name=collection_name)
retriever.insert_text_chunks(
    collection_name=collection_name,
    text_chunks=text_chunks,
    batch_size=512,
)
```
