```python
from glob import glob
from rich.pretty import pprint

from rage.loaders import PDFLoaeder
from rage.splitters import WordSplitter
from rage.retriever import Retriever
```

```python
loader = PDFLoaeder(disable_progress=True)
splitter = WordSplitter()
retriever = Retriever()
```

```python
file_paths = glob("/resources/documents/nietzsche/*.pdf")
print(f"file_paths => {len(file_paths)}")
```

```python
documents = await loader.batch_load(source_paths=file_paths)
text_chunks = splitter.split_documents(documents=documents)

print(f"text_chunks => {len(text_chunks)}")
```

```python
pprint(text_chunks[100])
```

```python
collection_name = "nietzsche"
retriever.create_collection(collection_name=collection_name)
retriever.insert_text_chunks(
    collection_name=collection_name, text_chunks=text_chunks
)
```
