```python
from glob import glob
from tqdm import tqdm
from more_itertools import flatten

from rage.loaders import PDFLoaeder
from rage.splitters import WordTextSplitter
from rage.retriever import Retriever
```

```python
loader = PDFLoaeder()
splitter = WordTextSplitter()
retriever = Retriever()
```

```python
file_paths = glob("/resources/documents/nietzsche/*.pdf")
print(f"file_paths => {len(file_paths)}")
```

```python
documents = [
    await loader.load(source_path=file_path) for file_path in tqdm(file_paths)
]

text_chunks = (splitter.split_documents(documents=docs) for docs in documents)
text_chunks = list(flatten(text_chunks))

print(f"text_chunks => {len(text_chunks)}")
```

```python
collection_name = "nietzsche"
retriever.create_collection(collection_name=collection_name)
retriever.insert_text_chunks(
    collection_name=collection_name, text_chunks=text_chunks
)
```
