```python
from rage.loaders import PDFLoaeder
from rage.splitters import SimpleTextSplitter
from rage.retriever import Retriever
```

```python
loader = PDFLoaeder()
splitter = SimpleTextSplitter()
retriever = Retriever(embeddings_show_progress_bar=True)
```

```python
documents = loader.load(
    source_path="/resources/documents/jung/el-libro-rojo-jung.pdf"
)

text_chunks = splitter.split_documents(documents=documents)
print(f"text_chunks => {len(text_chunks)}")
```

```python
collection_name = "el-libro-rojo"
retriever.create_collection(collection_name=collection_name)
retriever.insert_text_chunks(
    collection_name=collection_name, text_chunks=text_chunks
)
```
