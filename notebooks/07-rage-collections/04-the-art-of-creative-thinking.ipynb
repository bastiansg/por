```python
from rich.pretty import pprint

from rage.loaders import EpubLoader
from rage.splitters import TitleSplitter, WordSplitter
from rage.retriever import Retriever
```

```python
loader = EpubLoader()
splitter = WordSplitter()
retriever = Retriever()
```

```python
documents = await loader.load(
    source_path="/resources/documents/rod-judkins/el-arte-del-pensamiento-creativo.epub"
)
```

```python
title_splitter = TitleSplitter()
title_text_chunks = title_splitter.split_documents(documents=documents)
print(f"title_text_chunks => {len(title_text_chunks)}")
pprint(title_text_chunks[10])
```

```python
word_splitter = WordSplitter()
word_text_chunks = word_splitter.split_documents(documents=title_text_chunks)
print(f"word_text_chunks => {len(word_text_chunks)}")
pprint(word_text_chunks[10])
```

```python
collection_name = "el-arte-del-pensamiento-creativo"
retriever.create_collection(collection_name=collection_name)
retriever.insert_text_chunks(
    collection_name=collection_name,
    text_chunks=word_text_chunks,
)
```
