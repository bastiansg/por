```python
import cairosvg

from glob import glob
from tempfile import NamedTemporaryFile

from io import BytesIO
from PIL import Image, ImageOps
from por.multi_agent.nodes.utils import get_printer
```

```python
questions = [
    "What would cities look like if they had been designed for dreams instead of productivity?",
    "What would life be like if memory were collective rather than individual?",
    "What would civilization be like if time were not measured in hours?",
    "What would the present feel like if every decision had to consider the next seven generations?",
    "How would the world look if silence were treated as a shared public resource?",
]
```

```python
image_paths = sorted(glob("/resources/other-presents/*.svg"))
image_paths
```

```python
for image_path, question in zip(image_paths, questions):
    png_bytes = cairosvg.svg2png(url=image_path)
    image = Image.open(BytesIO(png_bytes)).convert("L")
    image = image.point(lambda p: 255 if p >= 250 else p)
    image = image.point(lambda p: 0 if p <= 200 else p)

    new_width = 576
    new_height = round(image.height * new_width / image.width)
    image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
    # break

    with NamedTemporaryFile(suffix=".png") as f:
        image.save(f, format="PNG")
        f.seek(0)

        printer = get_printer()
        printer.set(bold=True, align="center")
        printer.block_text(question)
        printer.set(bold=False)
        printer.text("\n\n")
        printer.text("------------------------------------------------")
        printer.text("\n\n")
        printer.image(img_source=f.name)
        printer.cut()
        printer.close()

    # break
```
