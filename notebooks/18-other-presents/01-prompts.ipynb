```python
import io
import base64

from PIL import Image
from openai import AsyncOpenAI

from rich.pretty import pprint
from rich.markdown import Markdown

from por.llm_agents import OracleImagePrompter, OracleImagePrompterDeps
```

```python
questions = [
    # "What would the world be like if money had never existed?",
    # "What would the present look like if land could not be owned?",
    # "How would society work if labor did not define a person's worth?",
    "What would cities look like if they had been designed for dreams instead of productivity?",
    "What would life be like if memory were collective rather than individual?",
    # "How would the world change if no language had words for ownership?",
    # "What would technology look like if it had evolved to listen instead of control?",
    "What would civilization be like if time were not measured in hours?",
    "What would the present feel like if every decision had to consider the next seven generations?",
    "How would the world look if silence were treated as a shared public resource?",
]

questions = [
    {
        "idx": idx,
        "question": q,
    }
    for idx, q in enumerate(questions, start=1)
]

print(questions)
```

```python
agent_deps_list = [
    OracleImagePrompterDeps(question=q["question"]) for q in questions
]
user_prompts = [
    "Provide the surreal image-generation prompt." for _ in questions
]

pprint(agent_deps_list)
pprint(user_prompts)
```

```python
oip = OracleImagePrompter()
oip_outputs = await oip.batch_generate(
    user_prompts=user_prompts,
    agent_deps_list=agent_deps_list,
)
```

```python
for question, oip_output in zip(questions, oip_outputs):
    pprint(question)
    display(Markdown(oip_output.flux_prompt))
```
